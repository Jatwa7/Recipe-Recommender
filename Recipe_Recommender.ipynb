```python
# Create a DataFrame
df = pd.DataFrame(data)

# Remove duplicate recipes
df.drop_duplicates(inplace=True)
```

```python
# Create TF-IDF matrix
tfidf = TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf.fit_transform(df["Ingredients"])
```

```python
print(tfidf.get_feature_names_out())
```

```python
# Function to recommend recipes based on selected ingredients
def recommend_recipes(selected_ingredients):

    recommended_recipes =[]

    # Convert selected ingredients to TF-IDF representation
    selected_ingredients_tfidf = tfidf.transform([" ".join(selected_ingredients)])
    #print(selected_ingredients_tfidf)

    # Compute cosine similarity between selected ingredients and all recipes
    cosine_similarities = linear_kernel(selected_ingredients_tfidf, tfidf_matrix).flatten()
    print(cosine_similarities)

    y=0
    for weight in cosine_similarities.tolist():
      if weight > 0.4:
        recommended_recipes.append(df['Recipe Name'][y])
      y=y+1


    return recommended_recipes
```
